## Analytical section

I chose to focus on the **freesurfer** data set. The data set includes measurements of volumes of different brain regions.

Each measurement is related to a **participant** with the code OAS3XXXX and a **session ID**  with the code OAS3XXXX_Freesurfer53_dYYYY. Where YYYY represents the number of days since the first entry. This info in addition to the age at entry was used to calculate the age of participants at the time of measurement.

In addition, the **Clinical Dementia Rating (CDR)** from the **ADRCCLINICAL DATA** was used to classify participants into cognitively normal participants or participants with Alzheimer disease. The criteria was a non-changing CDR score of zero across measurements.

Finally, the data set **subjects** was used to extract the gender of participants.

Mainly, the data can be further arranged to express the changes in time per participant. And it can also include other measurements of cognitive ability as inputs. 


The 2D plots show a decrease in brain volume with age, and with AD diagnosis. In addition, it shows an increase in Intracranial Volume. Both findings relate to known effects of age and AD diagnosis.
